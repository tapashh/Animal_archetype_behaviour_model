Run the cells top to bottom with **Shift+Enter**.

```python
import sys
sys.path.append("..")  # lets Jupyter find the animal_archetype package one folder up

from animal_archetype import PossibilisticArchetypeModel, TRAIT_NAMES, TRAITS
from animal_archetype.archetypes import ARCHETYPE_CENTROIDS

print(f"Loaded {len(TRAIT_NAMES)} traits and {len(ARCHETYPE_CENTROIDS)} archetypes.")
```

## 1. Build a trait vector for a person

Order matters here - it must follow `TRAIT_NAMES`. Run the next cell to see that order first.

```python
for i, t in enumerate(TRAITS):
    print(f"{i:2d}. {t.name:<20} ({t.category})")
```

```python
# Example: a made-up person's self-rated traits (0-100), in the exact order above
my_traits = [
    70, 65, 55, 80,   # strategic_thinking, learning_speed, creativity, curiosity
    60, 55, 40, 65,   # empathy, agreeableness, social_dependence, communication_style
    70,               # emotional_stability
    50, 75, 60,       # dominance, independence, long_term_planning
    35, 70, 80,       # impulsiveness, discipline, adaptability
    60, 65, 75,       # risk_taking, ambition, persistence
]

assert len(my_traits) == len(TRAIT_NAMES), "Vector length must match number of traits!"
```

## 2. Run the model

```python
model = PossibilisticArchetypeModel(beta=2500.0)
profile = model.profile(my_traits)

for archetype, score in profile.items():
    print(f"{archetype:<12} {score:6.2f}%")
```

## 3. Visualize the profile as a bar chart

```python
import matplotlib.pyplot as plt

names = list(profile.keys())
scores = list(profile.values())

plt.figure(figsize=(8, 4))
plt.barh(names[::-1], scores[::-1], color="#C97A2B")
plt.xlabel("Match percentage (%)")
plt.title("Archetype Profile")
plt.xlim(0, 100)
plt.tight_layout()
plt.show()
```

## 4. Experiment with the `beta` tuning parameter

This is the exact experiment your Data Scientist collaborators would run: see how strict vs. lenient the matching gets as `beta` changes.

```python
for beta_value in [500, 1000, 2500, 5000, 10000]:
    test_model = PossibilisticArchetypeModel(beta=beta_value)
    result = test_model.profile(my_traits)
    top_name, top_score = next(iter(result.items()))
    spread = max(result.values()) - min(result.values())
    print(f"beta={beta_value:<6} top match: {top_name} ({top_score:.1f}%)   spread across all archetypes: {spread:.1f} points")
```
